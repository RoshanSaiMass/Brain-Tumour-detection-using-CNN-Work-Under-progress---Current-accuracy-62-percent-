```python
import keras
import tensorflow
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization,Activation,GlobalAveragePooling2D)
from tensorflow.keras.optimizers import Adam, SGD
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import matplotlib.pyplot as plt
```

```python
train_datagen=ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    rotation_range=15
)
```

```python
test_datagen=ImageDataGenerator(
    rescale=1./255
)
```

```python
train_generator = train_datagen.flow_from_directory(
    "/content/Training",
    target_size=(224,224),
    batch_size=32,
    class_mode="sparse",
    color_mode="grayscale"
)
```

```python
test_generator=test_datagen.flow_from_directory(
    "/content/Testing",
    target_size=(224,224),
    batch_size=32,
    class_mode="sparse",
    color_mode="grayscale"
)
```

```python
print(train_generator.image_shape)
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy',
    min_delta=0,
    verbose=0,
    mode='auto',
    patience=5,
    restore_best_weights=False,
    baseline=None
)
```

```python
#Model 2
# Block 1
model2=Sequential()

model2.add(Conv2D(64, kernel_size=(7,7), padding='same',
                 input_shape=(224,224,1),activation='relu'))
model2.add(BatchNormalization())

model2.add(MaxPooling2D(pool_size=(2,2), strides=2))


# Block 2
model2.add(Conv2D(64, kernel_size=(3,3), padding='same',activation='relu'))
model2.add(BatchNormalization())

model2.add(MaxPooling2D(pool_size=(2,2), strides=2))


# Block 3
model2.add(Conv2D(128, kernel_size=(3,3), padding='same',activation='relu'))
model2.add(BatchNormalization())


model2.add(MaxPooling2D(pool_size=(2,2), strides=2))

#Block 4
model2.add(Conv2D(256, kernel_size=(3,3), padding='valid',activation='relu'))
model2.add(BatchNormalization())


model2.add(MaxPooling2D(pool_size=(2,2), strides=2))
#Block 5
model2.add(Conv2D(512, kernel_size=(3,3), padding='valid',activation='relu'))
model2.add(BatchNormalization())


model2.add(MaxPooling2D(pool_size=(2,2), strides=2))



model2.add(Flatten())

model2.add(Dense(
    1024,
    kernel_regularizer=tf.keras.regularizers.l2(0.01),
    kernel_initializer='he_uniform',activation='elu'
))
model2.add(BatchNormalization())
model2.add(Dropout(0.4))

model2.add(Dense(
    512,
    kernel_regularizer=tf.keras.regularizers.l2(0.01),
    kernel_initializer='he_uniform',activation='elu'
))
model2.add(BatchNormalization())
model2.add(Dropout(0.3))

model2.add(Dense(
    128,
    kernel_regularizer=tf.keras.regularizers.l2(0.01),
    kernel_initializer='he_uniform',activation='elu'
))
model2.add(BatchNormalization())
model2.add(Dropout(0.2))

model2.add(Dense(4, activation='softmax'))
```

```python
optimizer=Adam(learning_rate=1e-3)
model2.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history2=model2.fit(train_generator,epochs=50,validation_data=test_generator, batch_size=32)
```
